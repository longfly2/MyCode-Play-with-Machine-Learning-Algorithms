```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.uniform(-3, 3, size=100)
X = x.reshape(-1, 1)
y = 0.5 * x**2 + x + 2 + np.random.normal(0, 1, size=100)

plt.scatter(X, y)
plt.show()

```

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree=2)
pf.fit(X)
X2 = pf.transform(X)
print(X2[:5, :])
X[:5, :]
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X2, y)
print(lr.coef_)
print(lr.intercept_)

plt.scatter(x, y)
plt.plot(np.sort(x), lr.predict(X2)[np.argsort(x)], color='r')
plt.show()
```
